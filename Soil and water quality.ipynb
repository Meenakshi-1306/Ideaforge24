```python
import pandas as pd
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Load crop data
df = pd.read_csv(r"C:\Users\MANICKA MEENAKSHI.S\Downloads\Crop_recommendation.csv")

# Generate random water quality parameters
np.random.seed(42)  # For reproducibility
num_samples = df.shape[0]

# Define features including both crop and water quality parameters
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall', 
            'Dissolved Oxygen (mg/L)', 'Biochemical Oxygen Demand (mg/L)', 
            'Chemical Oxygen Demand (mg/L)', 'Total Dissolved Solids (mg/L)', 
            'Electrical Conductivity (μS/cm)', 'Coliform Bacteria (CFU/100mL)']

X = df[features]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans clustering
kmeans = KMeans(n_clusters=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Save the models
joblib.dump(kmeans, r"C:\Users\MANICKA MEENAKSHI.S\Downloads\kmeans_model.pkl")
joblib.dump(scaler, r"C:\Users\MANICKA MEENAKSHI.S\Downloads\scaler.pkl")

# Crop labels
crop_labels = np.array(['rice', 'maize', 'chickpea', 'kidneybeans', 'pigeonpeas',
                        'mothbeans', 'mungbean', 'blackgram', 'lentil', 'pomegranate'])

# Function to predict crop
def predict_crop(n, p, k, temperature, humidity, ph, rainfall, dissolved_oxygen, bod, cod, tds, ec, coliform):
    input_data = np.array([[n, p, k, temperature, humidity, ph, rainfall,
                             dissolved_oxygen, bod, cod, tds, ec, coliform]])
    input_scaled = scaler.transform(input_data)
    predicted_cluster = kmeans.predict(input_scaled)[0]
    predicted_crop = crop_labels[predicted_cluster]
    score = np.linalg.norm(input_scaled - kmeans.cluster_centers_[predicted_cluster])
    
    return predicted_crop, score

# Function to evaluate suitability based on score
def evaluate_suitability(score):
    if score < 0.5:
        return "Conditions are excellent for the predicted crop."
    elif score < 1.5:
        return "Conditions are suitable for the predicted crop."
    elif score < 2.5:
        return "Conditions are moderately suitable for the predicted crop."
    else:
        return "Conditions may not be optimal for the predicted crop."

# New input values


predicted_crop, score = predict_crop(n, p, k, temperature, humidity, ph, rainfall,
                                     dissolved_oxygen, bod, cod, tds, ec, coliform)
print(f"Predicted Crop: {predicted_crop}")
print(f"Score (Distance to Nearest Centroid): {score:.2f}")

suitability_message = evaluate_suitability(score)
print(suitability_message)
```
